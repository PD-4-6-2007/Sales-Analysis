```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("sales_data.csv")
```

```python
print("Dataset Preview")
print(df.head())
```

```python
print("\n Missing Values")
print(df.isnull().sum())
```

```python
df=df.dropna()
```

```python
df=df.drop_duplicates()
```

```python
print(df.columns)
```

```python
total_revenue=df["Sales_Amount"].sum()
print("Total Revenue=",total_revenue)
```

```python
top_products=df.groupby("Product_Category")["Sales_Amount"].sum().sort_values(ascending=False)
print("\n Top selling Products")
print(top_products.head(5))
```

```python
region_sales=df.groupby("Region")["Sales_Amount"].sum()
region_sales.plot(kind="bar")
plt.title("Sales By Region")
plt.xlabel("Region")
plt.ylabel("Sales_Amount")
plt.show()
```

```python
avg_discount=df["Discount"].mean()
print("Average Discount=",avg_discount)
```

```python
region_sales=df.groupby("Region")["Sales_Amount"].sum()
print(region_sales)
```

```python
channel_sales=df.groupby("Sales_Channel")["Sales_Amount"].sum()
print(channel_sales)
```

```python
category_sales=df.groupby("Product_Category")["Sales_Amount"].sum()
print(category_sales)
```

```python
payment_sales=df.groupby("Payment_Method")["Sales_Amount"].sum()
print(payment_sales)
```

```python
customer_sales=df.groupby("Customer_Type")["Sales_Amount"].sum()
print(customer_sales)
```

```python
import matplotlib.pyplot as plt
category_sales=df.groupby("Product_Category")["Sales_Amount"].sum()
category_sales.plot(kind="line",marker="o")
plt.title("Sales by Product Category")
plt.xlabel("Product_Category")
plt.ylabel("Sales_Amount")
plt.show()
```

```python
import matplotlib.pyplot as plt 
plt.hist(df['Sales_Amount'],bins=10)
plt.title('Sales Amount Distribution')
plt.xlabel('Sales Amount')
plt.ylabel('Fequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.boxplot(df['Sales_Amount'])
plt.title('Sales Amount Box Plot')
plt.show()
```

```python
monthly_sales=df.groupby('Sale_Date')['Sales_Amount'].sum()
monthly_sales.plot(kind='area')
plt.title('Area Chart of Sales')
plt.show()
```

```python
region_sales=df.groupby('Region')['Sales_Amount'].sum()
region_sales.plot(kind='barh')
plt.title('Sales by Region')
plt.show()
```

```python
category_sales=df.groupby('Product_Category')['Sales_Amount'].sum()
plt.pie(category_sales,labels=category_sales.index,autopct='%1.1f%%')
centre_circle=plt.Circle((0,0),0.70,fc='white')
fig=plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Product Category Sales')
plt.show()
```
